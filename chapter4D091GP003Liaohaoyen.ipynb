#### Spatial Filters

```python
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image using cv2.
a = cv2.imread('/content/ultrasound_muscle.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)

# Initializing the filter of size 5 by 5.
# The filter is divided by 25 for normalization.
k = np.ones((5,5))/25 # The filter is divided by 25 for normalization.
# performing convolution
g = np.ones((5,5))/500 # The filter is divided by 500 for normalization.
b = scipy.ndimage.filters.convolve(a, k)
# Writing b to a file.
c = scipy.ndimage.filters.convolve(a, g)
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
```

#### Median Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('/content/ct_saltandpepper.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing the median filter.
b = scipy.ndimage.filters.median_filter(a, size=5)
# Saving b as median_output.png in Figures folder
c = scipy.ndimage.filters.median_filter(a, size=15)
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
```

#### Sobel Filter

```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('/content/prewitt_cir.png')
e = cv2.imread('/content/prewitt_cir.png')
# Converting a to grayscale .
e = cv2.cvtColor(e, cv2.COLOR_BGR2XYZ)
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing Sobel filter.
b = ndimage.sobel(a)
# Saving b.
# Performing Sobel filter.
d = ndimage.sobel(e)
# Saving d.
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(d)
plt.show()
```

#### Canny Filter

```python
import cv2
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('/content/maps1.png')
# Performing Canny edge filter.
b = cv2.Canny(a, 100, 200)
# Saving b.
c = cv2.Canny(a, 500, 200)
# Saving c.
d = cv2.Canny(a, 100, 10)
# Saving d.
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
plt.imshow(d)
plt.show()
```

#### Laplacian Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('/content/imagefor_laplacian.png')
# Performing Laplacian filter.
b = scipy.ndimage.filters.laplace(a,mode='reflect')
# Performing Laplacian filter.
c = scipy.ndimage.filters.laplace(a,mode='mirror')
plt.imshow(a)
plt.show()
plt.imshow(b)
plt.show()
plt.imshow(c)
plt.show()
```

# Laplacian of Gaussian Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('/content/vhuman_t1.png')
# Performing Laplacian of Gaussian.
b = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='reflect')
c = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='constant')
plt.imshow(a) 
plt.show()
plt.imshow(b) 
plt.show()
plt.imshow(c) 
plt.show()
```

# Frangi Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
from PIL import Image
from skimage.filters import frangi

img = cv2.imread('/content/angiogram1.png')
img1 = np.asarray(img)
img2 = frangi(img1, black_ridges=True)
img3 = 255*(img2-np.min(img2))/(np.max(img2)-np.min(img2))
img4 = 4*(img2-np.min(img2))/(np.max(img2)-np.min(img2))
plt.imshow(img)
plt.show()
plt.imshow(img2)
plt.show()
plt.imshow(img3)
plt.show()
plt.imshow(img4)
```
