## dilation

```python
from PIL import Image
import scipy.ndimage as snd
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = Image.open('../figures/dil_image.png').convert('L')
a = np.array(a)
# Performing binary dilation for 5 iterations.
b = snd.morphology.binary_dilation(a, iterations=100)
# Saving the image as 8-bit as b is a
# binary image of dtype=bool
# img= Image.open('./test/test2.bmp')
# plt.imshow(a,cmap='gray',vmin=0,vmax=255)
# cv2.imshow("oke", b)
# cv2.waitKey(0)
# fig = plt.figure()
# ax = fig.add_subplot(1, 2, 1)
# imgplot = plt.imshow(a, cmap='gray',vmin=0,vmax=255)
# ax.set_title('Before')
# ax = fig.add_subplot(1, 2, 2)
# imgplot = plt.imshow(b, cmap='gray')
# ax.set_title('After')
# plt.show()



rows = 1
cols = 7
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a,cmap='gray')
        plt.axis('off')
    else:
        b = snd.morphology.binary_dilation(a, iterations=10*i)
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("iteration: " +str(10*i))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b,cmap='gray')
        plt.axis("off")
fig.tight_layout()    
plt.show()
```

## grayscale dilation erosion

```python
import numpy as np
from PIL import Image
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it into grayscale.
a = Image.open('../figures/sem3.png').convert('L')
# Creating a structuring element.
footprint = np.ones((15,15))
# Performing grey dilation.
b = scipy.ndimage.morphology.grey_dilation(a, footprint=footprint)
# Converting ndarray to image.
c = Image.fromarray(b)
# Saving the image.
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, cmap='gray')
ax.set_title('Before')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, cmap='gray')
ax.set_title('After')
plt.show()
```

## opening and closing

```python
from PIL import Image
import scipy.ndimage as snd
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Opening the image and converting it to .
a = Image.open('../figures/dil_image.png').convert('L')
a = np.array(a)
# Defining the structuring element.
s = [[0,1,0],[1,1,1], [0,1,0]]
# Performing the binary opening for 5 iterations.
b = snd.morphology.binary_opening(a, structure=s, iterations=5)
# Saving the image as 8-bit as b is a
# binary image of dtype=bool
# fig = plt.figure()
# ax = fig.add_subplot(1, 2, 1)
# imgplot = plt.imshow(a, cmap='gray')
# ax.set_title('Before')
# ax = fig.add_subplot(1, 2, 2)
# imgplot = plt.imshow(b, cmap='gray')
# ax.set_title('After')
# plt.show()


rows = 1
cols = 7
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a,cmap='gray')
        plt.axis('off')
    else:
        b = snd.morphology.binary_opening(a, structure=s, iterations=5*i)
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("iteration: " +str(5*i))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b,cmap='gray')
        plt.axis("off")
fig.tight_layout()    
plt.show()
```

## grayscale opening and closing

```python

import numpy as np
from PIL import Image
import scipy.ndimage 
import matplotlib.pyplot as plt

# Opening the image and converting it into grayscale.
a = Image.open('../figures/adaptive_example1.png').convert('L')
# Creating a structuring element.
footprint = np.ones((40,40))
# Performing grey opening.
b = scipy.ndimage.morphology.grey_opening(a, footprint=footprint)
# Converting ndarray to image.
c = Image.fromarray(b)
# Saving the image.
c.save('../figures/grey_opening_output_40.png') 
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, cmap='gray')
ax.set_title('Before')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, cmap='gray')
ax.set_title('After')
plt.show()
```

```python
import numpy as np
from PIL import Image
from skimage.morphology import skeletonize
import cv2
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = Image.open('../figures//steps1.png').convert('L')
# Converting a to an ndarray and normalizing it.
a = np.asarray(a)/np.max(a)
# Performing skeletonization.
b = skeletonize(a)
# Saving the image as 8-bit as b is a
# binary image of dtype=bool
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, cmap='gray')
ax.set_title('Before')
ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, cmap='gray')
ax.set_title('After')
plt.show()
```
