# rotation

```python
import numpy as np
import scipy.misc,math
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = cv2.imread('../Figures/angiogram1.png')
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


# rotation angle in radians
transformation = AffineTransform(rotation=0.05)
img2 = warp(img1, transformation)
# img4 = scipy.misc.toimage(img2).convert('LA')
#plt.imshow(img2,cmap='gray')
#plt.show()

rows = 1
cols = 5
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(img1,cmap='gray')
    else:
        transformation = AffineTransform(rotation=0.1*i)
        # nearest neighbor order = 0
        img2 = warp(img1, transformation, mode='constant')
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("x,y translation: "+str(round(0.1*i,2)) + " radian")
        axes[-1].set_title(subplot_title)  
        plt.imshow(img2,cmap='gray')
        plt.axis("off")
fig.tight_layout()    
plt.show()
```

# scaling

```python
import cv2
import numpy as np
import scipy.misc, math
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = cv2.imread('../Figures/angiogram1.png')
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#scale by1/2 on both x and y.
transformation = AffineTransform(scale= (0.5, 0.5))
img2 = warp(img1, transformation)
#plt.imshow(img2,cmap='gray')
#plt.show()

rows = 1
cols = 5
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(img1,cmap='gray')
    else:
        transformation = AffineTransform(scale= (0.5*i, 2*i))
        # nearest neighbor order = 0
        img2 = warp(img1, transformation, mode='constant')
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("(x,y) magnification: (" + str(0.5*i) + "," + str(2*i) + ")")
        axes[-1].set_title(subplot_title)  
        plt.imshow(img2,cmap='gray')
        plt.axis("off")
fig.tight_layout()    
plt.show()
```

```python
import numpy as np
import scipy.misc, math
import cv2
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = cv2.imread('../Figures/angiogram1.png')
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

transformation = AffineTransform(scale=(0.3, 0.3))


rows = 1
cols = 5
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 4:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(img1,cmap='gray')
    else:
        # nearest neighbor order = 0
        img2 = warp(img1, transformation, order=i)
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("order: " + str(i))
        axes[-1].set_title(subplot_title)  
        plt.imshow(img2,cmap='gray')
        plt.axis("off")
fig.tight_layout()    
plt.show()



```
