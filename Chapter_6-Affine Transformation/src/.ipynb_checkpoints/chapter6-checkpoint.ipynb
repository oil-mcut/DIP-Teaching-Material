```python
#comment This code is modified as the privious code was not working 

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../Figures/angiogram1.png',0)
plt.imshow(img)
plt.title("original colour image")
plt.show()

rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst)
plt.title("Translation image")
plt.show()

M = np.float32([[1,0,200],[0,1,150]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
plt.title("Translation image")
plt.show()

#comment we can controll the tranlation according to our need by increasing or decreasing the value og the coordinates 

```

#### 6.2.2 Rotation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('../Figures/angiogram1.png',0)
plt.imshow(img)
plt.title("original colour image")
plt.show()
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst)
plt.title("90 degree rotation image")
plt.show()

M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst)
plt.title("180 degree rotation image")
plt.show()

M = cv2.getRotationMatrix2D((cols/2,rows/2),150,1)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.imshow(dst)
plt.title("150 degree rotation image")
plt.show()

# comment rotation can be controlled by changing the rotation angle. 
```

#### 6.2.3 Scaling

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../Figures/angiogram1.png')
plt.imshow(img)
plt.show()

res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
plt.show()

#OR

height, width = img.shape[:2]
res = cv2.resize(img,(4*width, 4*height), interpolation = cv2.INTER_CUBIC)
plt.imshow(res)
plt.show()

  # comment one can enlarge or reduce the size of image 
```

#### 6.2.4 Interpolation

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/angiogram1.png')
img1 = scipy.misc.fromimage(img)

transformation = AffineTransform(scale=(0.3, 0.3))

# nearest neighbor order = 0
img2 = warp(img1, transformation, order=0)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

# bi-linear order = 1
img2 = warp(img1, transformation, order=1) # default
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#bi-quadratic order = 2
img2 = warp(img1, transformation, order=2)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#bi-cubic order = 3
img2 = warp(img1, transformation, order=3)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#comment when it comes to increasing or decreasing the size of image pixel size do not get bigger or smaller but more pixels are addesd
# or subtracted.

# different type of interpolation method are used to fill the newly generated pixels.
```
