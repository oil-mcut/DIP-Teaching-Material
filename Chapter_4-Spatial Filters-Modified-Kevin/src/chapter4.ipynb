#### Spatial Filters
#### i try change the mask matrix becomes 20x20 and the filter become larger, so the result will become zoom

```python
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image using cv2.
a = cv2.imread('../Figures/ultrasound_muscle.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)

# Initializing the filter of size 5 by 5.
# The filter is divided by 25 for normalization.
k = np.ones((20,20))/400
# performing convolution
b = scipy.ndimage.filters.convolve(a, k)
# Writing b to a file.
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(np.rot90(a,2))
f.add_subplot(1,2, 2)
plt.imshow(np.rot90(b,2))
plt.show(block=True)
```

#### Median Filter
### i try to make the filter size of median filter become 100, and the result is like picture 3

```python
import cv2
import scipy.ndimage
import numpy as np
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('../Figures/ct_saltandpepper.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing the median filter.
b = scipy.ndimage.filters.median_filter(a, size=10)
c = scipy.ndimage.filters.median_filter(a, size=100)
# Saving b as median_output.png in Figures folder
f = plt.figure()
f.add_subplot(1,3, 1)
plt.imshow(np.rot90(a,2))
f.add_subplot(1,3, 2)
plt.imshow(np.rot90(b,2))
f.add_subplot(1,3, 3)
plt.imshow(np.rot90(c,2))
plt.show(block=True)
```

#### Max Filter 
### i modify the size of maximum filter into 100, and the result in 2nd picture

```python
import scipy.misc
import scipy.ndimage
from scipy.misc.pilutil import Image
import matplotlib.pyplot as plt
import numpy as np


# opening the image and converting it to grayscale
a = Image.open('../Figures/wave.png').convert('L')
# performing maximum filter
b = scipy.ndimage.filters.maximum_filter(a, size=5)
c = scipy.ndimage.filters.maximum_filter(a, size=100)
# b is converted from an ndarray to an image
b = scipy.misc.toimage(b)
c = scipy.misc.toimage(c)
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(np.rot90(b,2))
f.add_subplot(1,2, 2)
plt.imshow(np.rot90(c,2))
plt.show(block=True)
```

#### Sobel Filter
### i modify this sobel filter with show the real picture and the result after sobel filter in picture 2

```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np

# Opening the image.
a = cv2.imread('../Figures/cir.png')
# Converting a to grayscale .
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing Sobel filter.
b = ndimage.sobel(a)
# Saving b.
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(np.rot90(a,2))
f.add_subplot(1,2, 2)
plt.imshow(np.rot90(b,2))
plt.show(block=True)
```

#### Laplacian Filter
### i modify the mode of filter becoma "constant" and i show the real image and the result after filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt
import numpy as np
# Opening the image.
a = cv2.imread('../Figures/imagefor_laplacian.png')
# Performing Laplacian filter.
b = scipy.ndimage.filters.laplace(a,mode='constant')
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(np.rot90(a,2))
f.add_subplot(1,2, 2)
plt.imshow(np.rot90(b,2))
plt.show(block=True)
```

#### Laplacian of Gaussian Filter
### i modify the mode of filter becoma "constant" and i show the real image and the result after filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt
import numpy as np

# Opening the image.
a = cv2.imread('../Figures/vhuman_t1.png')
# Performing Laplacian of Gaussian.
b = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='reflect')
c = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='constant')
f = plt.figure()

f.add_subplot(1,3, 1)
plt.imshow(np.rot90(a,2))
f.add_subplot(1,3, 2)
plt.imshow(np.rot90(b,2))
f.add_subplot(1,3, 3)
plt.imshow(np.rot90(c,2))
plt.show(block=True)
```

#### Frangi Filter
### i show the result before and after filter in below

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
from PIL import Image
from skimage.filters import frangi

img = cv2.imread('../Figures/angiogram1.png')
img1 = np.asarray(img)
img2 = frangi(img1, black_ridges=True)
img3 = 255*(img2-np.min(img2))/(np.max(img2)-np.min(img2))
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(np.rot90(img,2))
f.add_subplot(1,2, 2)
plt.imshow(np.rot90(img3,2))
plt.show(block=True)
```
