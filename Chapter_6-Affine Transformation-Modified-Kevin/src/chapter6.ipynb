### Translation
### i try flower image for this transformation and traslate by (200,100) x,y pixel

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/test.jpg').convert('L')
img1 = scipy.misc.fromimage(img)

# translate by 10 pixels in x and 4 pixels in y

transformation = AffineTransform(translation=(200, 100))
# nearest neighbor order = 0
img2 = warp(img1, transformation)
im4 = scipy.misc.toimage(img2).convert('LA')
# img = 
plt.imshow(im4)
plt.show()
```

#### 6.2.3 Scaling
### i try image flower and i scale the image by 0,5 and 0,5 for x and y

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/test.jpg').convert('L')
img1 = scipy.misc.fromimage(img)

#scale by1/2 on both x and y.
transformation = AffineTransform(scale= (0.5, 0.5))
img2 = warp(img1, transformation)
im4 = scipy.misc.toimage(img2).convert('LA')
plt.imshow(im4)
plt.show()
```

#### 6.2.4 Interpolation

```python
import numpy as np
import scipy.misc, math
from scipy.misc.pilutil import Image
from skimage.transform import AffineTransform, warp
import matplotlib.pyplot as plt

img = Image.open('../Figures/test.jpg')
img1 = scipy.misc.fromimage(img)

transformation = AffineTransform(scale=(1, 1))

# nearest neighbor order = 0
img2 = warp(img1, transformation, order=0)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

# bi-linear order = 1
img2 = warp(img1, transformation, order=1) # default
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#bi-quadratic order = 2
img2 = warp(img1, transformation, order=2)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()

#bi-cubic order = 3
img2 = warp(img1, transformation, order=3)
im4 = scipy.misc.toimage(img2)
plt.imshow(im4)
plt.show()
```
