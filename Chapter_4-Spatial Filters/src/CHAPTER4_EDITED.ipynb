#### SPATIAL FILTERS

```python
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image using cv2.
a = cv2.imread('../Figures/ultrasound_muscle.png')

# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)

# Initializing the filter of size 3 by 3.
# The filter is divided by 25 for normalization.
k1 = np.ones((3,3))/9
k2 = np.ones((30,30))/900
# performing convolution
b33 = scipy.ndimage.filters.convolve(a, k1)
b3030 = scipy.ndimage.filters.convolve(a, k2)
# Writing b to a file.
plt.imshow(b33)

#plt.show()
```

```python
plt.imshow(b3030)
```

## note for spatial filter:
## the higher dimension of window matrix, it convolved bigger area, so the result will be more blur and uniform

### Median Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('../Figures/ct_saltandpepper.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing the median filter.
# Saving b as median_output.png in Figures folder
rows = 1
cols = 3
axes=[]
fig=plt.figure()
for i in range(rows*cols):
    b = scipy.ndimage.filters.median_filter(a, size=2*5**(i))
    axes.append( fig.add_subplot(rows, cols, i+1) )
    subplot_title=("window size "+str(2*5**i))
    axes[-1].set_title(subplot_title)  
    plt.imshow(b)
fig.tight_layout()    
plt.show()
```

```python
plt.title("Original image")
plt.imshow(a)
```

## as the window size of median filter increase, the salt and pepper noise will be decrease.
## but the contrast will be brighter as the window size increase

### max filter

```python
import scipy.misc
import scipy.ndimage
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
# Opening the image using cv2.
a = cv2.imread('../Figures/wave.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# performing maximum filter
b = scipy.ndimage.filters.maximum_filter(a, size=5)
# b is converted from an ndarray to an image

rows = 1
cols = 4
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a)
    else:
        b = scipy.ndimage.filters.maximum_filter(a, size=2*5**(i-1))
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("window size "+str(2*5**(i-1)))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b)
fig.tight_layout()    
plt.show()




#plt.imshow(a)
#plt.show()
```

## when the window size of maximum filter increase, image will be shifted upside due to the increase of each pixel value in the image

```python
import scipy.misc
import scipy.ndimage
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
# Opening the image using cv2.
a = cv2.imread('../Figures/wave.png')
# Converting the image to grayscale.
#a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# performing maximum filter
#b = scipy.ndimage.filters.maximum_filter(a, size=5)
# b is converted from an ndarray to an image

rows = 1
cols = 4
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a)
    else:
        b = scipy.ndimage.filters.minimum_filter(a, size=2*5**(i-1))
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("window size "+str(2*5**(i-1)))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b)
fig.tight_layout()    
plt.show()
```

### sobel filter

```python
import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
# Opening the image using cv2.
a = cv2.imread('../Figures/cir.png')
# Converting the image to grayscale.
#a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# performing maximum filter
#b = scipy.ndimage.filters.maximum_filter(a, size=5)
# b is converted from an ndarray to an image

rows = 1
cols = 4
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 3:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a)
    else:
        b = ndimage.sobel(a,axis=i)
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("axis "+str(i))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b)
fig.tight_layout()    
plt.show()
```

## LAPLACIAN FILTER

```python
import scipy.misc
import scipy.ndimage
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
# Opening the image using cv2.
a = cv2.imread('../Figures/imagefor_laplacian.png')
# Converting the image to grayscale.
#a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# performing maximum filter
#b = scipy.ndimage.filters.maximum_filter(a, size=5)
# b is converted from an ndarray to an image
mod= {1:'reflect', 2:'constant', 3:'nearest', 4:'mirror', 5:'wrap'}
rows = 1
cols = 6
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a)
    else:
        b = scipy.ndimage.filters.laplace(a,mode=mod[i])
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("mode "+str(mod[i]))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b)
fig.tight_layout()    
plt.show()
```

## Laplacian of gaussian filter

```python
import scipy.misc
import scipy.ndimage
import matplotlib.pyplot as plt

# opening the image and converting it to grayscale
# Opening the image using cv2.
a = cv2.imread('../Figures/vhuman_t1.png')
# Converting the image to grayscale.
#a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# performing maximum filter
#b = scipy.ndimage.filters.maximum_filter(a, size=5)
# b is converted from an ndarray to an image
mod= {1:'reflect', 2:'constant', 3:'nearest', 4:'mirror', 5:'wrap'}
rows = 1
cols = 6
axes=[]
fig=plt.figure(figsize=(15, 15))
for i in range(rows*cols):
    if i == 0:
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("original image")
        axes[-1].set_title(subplot_title)  
        plt.imshow(a)
    else:
        b = scipy.ndimage.filters.gaussian_laplace(a,sigma=1,mode=mod[i])
        axes.append( fig.add_subplot(rows, cols, i+1) )
        subplot_title=("mode "+str(mod[i]))
        axes[-1].set_title(subplot_title)  
        plt.imshow(b)
fig.tight_layout()    
plt.show()
```
