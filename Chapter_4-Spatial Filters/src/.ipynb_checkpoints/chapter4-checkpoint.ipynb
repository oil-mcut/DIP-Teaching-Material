#### Spatial Filters

```python
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image using cv2.
a = cv2.imread('../Figures/ultrasound_muscle.png')
plt.imshow(a)
plt.title("gray scale Image")
plt.show()
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
plt.imshow(a)
plt.title("Colour Image")
plt.show()


# Initializing the filter of size 5 by 5.
# The filter is divided by 25 for normalization.
k = np.ones((5,5))/25
# performing convolution
b = scipy.ndimage.filters.convolve(a, k)
# Writing b to a file.
plt.imshow(b)
plt.title("Filter result 5*5")
plt.show()
l = np.ones((6,6))/36
# performing convolution
b = scipy.ndimage.filters.convolve(a, l)
# Writing b to a file.
plt.imshow(b)
plt.title("Filter result 6*6")
plt.show()
m = np.ones((10,10))/100
# performing convolution
b = scipy.ndimage.filters.convolve(a, m)
# Writing b to a file.
plt.imshow(b)
plt.title("Filter result 10*10")
plt.show()

#comment as the size of the mask increse the blure increase 
```

#### Median Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('../Figures/ct_saltandpepper.png')
plt.imshow(a)
plt.title("gray scale Image")
plt.show()
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
plt.imshow(a)
plt.title("colour Image")
plt.show()
# Performing the median filter.
b = scipy.ndimage.filters.median_filter(a, size=5)
# Saving b as median_output.png in Figures folder
plt.imshow(b)
plt.title("Filter image")
plt.show()
#The below code can give a batter filter then the above as comparable from the output.
import cv2
import numpy as np

a = cv2.imread('../Figures/ct_saltandpepper.png')
plt.imshow(a)
plt.title("image")
plt.show()
median = cv2.medianBlur(a, 5)
compare = np.concatenate((a, median), axis=1) #side by side comparison
plt.imshow(compare)
plt.title("compare image")
plt.show()

```

#### Sobel Filter

```python
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/cir.png')
plt.imshow(a)
plt.title("original image")
plt.show()
# Converting a to grayscale .
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
plt.imshow(a)
plt.title("colour image ")
plt.show()
# Performing Sobel filter.
b = ndimage.sobel(a)
# Saving b.
plt.imshow(b)
plt.title("sobel filter")
plt.show()
#In this code i tried to represent  the input and output of the sobel and se how sobel work 
#comparision can be done with different soble and other filter
```

#### Canny Filter

```python
import cv2
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/maps1.png')
plt.imshow(a)
plt.title("original image")
plt.show()
# Performing Canny edge filter.
b = cv2.Canny(a, 100, 200)
# Saving b.
plt.imshow(b)
plt.title("canny filter(100,200)")
plt.show()

b = cv2.Canny(a, 0, 0)
# Saving b.
plt.imshow(b)
plt.title("canny filter (0,0)")
plt.show()

b = cv2.Canny(a, 500, 1000)
# Saving b.
plt.imshow(b)
plt.title("canny filter (500,1000)")
plt.show()

#comment: In this code we see if we icrease the input value the sharpness of the letter increase 
#For 0,0 the image is blur and is not good.
# for higher value the edge is more clear
```

# Laplacian of Gaussian Filter

```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image.
a = cv2.imread('../Figures/vhuman_t1.png')
plt.imshow(a)
plt.title("original image")
plt.show()
# Performing Laplacian of Gaussian.
b = scipy.ndimage.filters.gaussian_laplace(a, sigma=1, mode='reflect')
plt.imshow(b)
plt.title("Sigma=1")
plt.show()

b = scipy.ndimage.filters.gaussian_laplace(a, sigma=2, mode='reflect')
plt.imshow(b)
plt.title("Sigma=2")
plt.show()

# comment: If we increase the sigma value we see for sigma=1 have some noise 
# when sigma is 2 this noise is removed 
```

# Frangi Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
from PIL import Image
from skimage.filters import frangi

img = cv2.imread('../Figures/angiogram1.png')
plt.imshow(img)
plt.title("img")
plt.show()

img1 = np.asarray(img)
plt.imshow(img1)
plt.title("img1")
plt.show()

img2 = frangi(img1, black_ridges=True)
plt.imshow(img2)
plt.title("img2")
plt.show()

img3 = 255*(img2-np.min(img2))/(np.max(img2)-np.min(img2))
plt.imshow(img3)
plt.title("img3")
plt.show()

# Comment: From the final image we can conclude that more information can be obtained from the final image then the original image 
```
